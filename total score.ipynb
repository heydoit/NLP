### 데이터 불러오기

```python
import pandas as pd
Lyrics = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/odatt/Lyrics_final.csv')
knu_pos = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/odatt/KUN_pos.csv', encoding = 'cp949')
knu_neg = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/odatt/KUN_neg.csv', encoding = 'cp949')
kosac_pos = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/odatt/kosac_pos.csv', encoding = 'cp949')
kosac_neg = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/odatt/kosac_neg.csv', encoding = 'cp949')
```

### 긍정 가사 데이터 합산

```python
# 행복, 희망 => 1(긍정), 슬픔, 분노 => 0(부정) 으로 이진분류
def rating_to_label(label):
    if label == 0 or label == 3:
        return 1
    else:
        return 0

Lyrics['y'] = Lyrics['label'].apply(lambda x: rating_to_label(x))
Lyrics_pos = Lyrics[Lyrics['y'] == 1]
Lyrics_neg = Lyrics[Lyrics['y'] == 0]

# pos score 모음
Lyrics_pos = Lyrics_pos.reset_index()
Lyrics_pos = Lyrics_pos.drop(['index'], axis = 1)
Lyrics_pos['knu_score'] = knu_pos['total_score']
Lyrics_pos['kosac_score'] = kosac_pos['total_score']
```

```python
s1_pos = Lyrics_pos['knu_score']
s2_pos = Lyrics_pos['kosac_score']
total_pos = s1_pos+s2_pos

Lyrics_pos['total'] = total_pos
```
