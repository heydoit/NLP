### 데이터 불러오기

```python
import pandas as pd
Lyrics = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/odatt/Lyrics_final.csv')
knu_pos = pd.read_csv('/content/KUN_pos.csv', encoding = 'cp949')
knu_neg = pd.read_csv('/content/KUN_neg.csv', encoding = 'cp949')
kosac_pos = pd.read_csv('/content/kosac_pos.csv', encoding = 'cp949')
kosac_neg = pd.read_csv('/content/kosac_neg.csv', encoding = 'cp949')
```

### 긍정 가사 데이터 합산

```python
# 행복, 희망 => 1(긍정), 슬픔, 분노 => 0(부정) 으로 이진분류
def rating_to_label(label):
    if label == 0 or label == 3:
        return 1
    else:
        return 0

Lyrics['y'] = Lyrics['label'].apply(lambda x: rating_to_label(x))
Lyrics_pos = Lyrics[Lyrics['y'] == 1]
Lyrics_neg = Lyrics[Lyrics['y'] == 0]

# # pos score 모음
# Lyrics_pos = Lyrics_pos.reset_index()
# Lyrics_pos = Lyrics_pos.drop(['index'], axis = 1)
# Lyrics_pos['knu_score'] = knu_pos['total_score']
# Lyrics_pos['kosac_score'] = kosac_pos['total_score']
```

```python
# 열 합계
print('knu_score 합계 : ', Lyrics_pos['knu_score'].sum())
print('kosac_score 합계 : ',Lyrics_pos['kosac_score'].sum())
print('total 합계 : ',Lyrics_pos['total'].sum())
```

### 부정 가사 데이터 합산

```python
# neg score 모음

# Lyrics_neg = Lyrics_neg.reset_index()
Lyrics_neg = Lyrics_neg.drop(['index'], axis = 1)
Lyrics_neg['knu_score'] = knu_neg['total_score']
Lyrics_neg['kosac_score'] = kosac_neg['total_score']
```

```python
s1_neg = Lyrics_neg['knu_score']
s2_neg = Lyrics_neg['kosac_score']
total_neg = s1_neg+s2_neg

Lyrics_neg['total'] = total_neg
```

```python
# 열 합계
print('knu_score 합계 : ', Lyrics_neg['knu_score'].sum())
print('kosac_score 합계 : ',Lyrics_neg['kosac_score'].sum())
print('total 합계 : ',Lyrics_neg['total'].sum())
```

### 총 데이터 합산

```python
total_score = pd.concat([Lyrics_pos, Lyrics_neg])
total_score = total_score.sort_values('label')
total_score = total_score.reset_index()
total_score = total_score.drop('index', axis = 1)
total_score
```

```python
total_score.to_csv("Total_score.csv", mode='w')
```

### knu, kosac

```python
# 긍부정 라벨링 

knu_pos.loc[knu_pos['pos_neg'] == '긍정', 'pos_neg'] = 1
knu_pos.loc[knu_pos['pos_neg'] == '부정', 'pos_neg'] = 0

knu_neg.loc[knu_neg['pos_neg'] == '긍정', 'pos_neg'] = 1
knu_neg.loc[knu_neg['pos_neg'] == '부정', 'pos_neg'] = 0

kosac_pos.loc[kosac_pos['pos_neg'] == '긍정', 'pos_neg'] = 1
kosac_pos.loc[kosac_pos['pos_neg'] == '부정', 'pos_neg'] = 0

kosac_neg.loc[kosac_neg['pos_neg'] == '긍정', 'pos_neg'] = 1
kosac_neg.loc[kosac_neg['pos_neg'] == '부정', 'pos_neg'] = 0
```

```python
# 필요한 데이터프레임 열만 추출 

knu_pos_col = knu_pos[['total_score','pos_neg']]
knu_neg_col = knu_neg[['total_score','pos_neg']]

kosac_pos_col = kosac_pos[['total_score','pos_neg']]
kosac_neg_col = kosac_neg[['total_score','pos_neg']]
```

```python
# knu 데이터화 및 csv 저장 
knu_df = pd.concat([knu_pos_col, knu_neg_col], axis = 0)
knu_df = knu_df.reset_index()
knu_df = knu_df.drop('index', axis = 1)
Lyrics_knu = pd.concat([Lyrics, knu_df], axis = 1)
Lyrics_knu.to_csv('Lyrics_Knu.csv')
```

```python
#kosac 데이터화 및 csv 저장 
kosac_df = pd.concat([kosac_pos_col, kosac_neg_col], axis = 0)
kosac_df = kosac_df.reset_index()
kosac_df = kosac_df.drop('index', axis = 1)

Lyrics_kosac = pd.concat([Lyrics, kosac_df], axis = 1)

Lyrics_kosac.to_csv('Lyrics_koSac.csv')
```
